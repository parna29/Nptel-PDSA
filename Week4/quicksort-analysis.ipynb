```python
#demonstrate that randomizing a list gives faster results than already sorted list
import random
def randomize(l):
    for i in range(len(l)//2):
        j = random.randrange(0, len(l), 1)
        k = random.randrange(0, len(l), 1)
        (l[j], l[k]) = (l[k], l[j])

def Quicksort(A, l, r): #sort A[l:r
    if r - l <= 1: #base case
        return ()

    #partition wrt pivot, a[l]
    yellow = l + 1

    for green in range(l+1, r):
        if A[green] <= A[l]:
            (A[yellow], A[green]) = (A[green], A[yellow])
            yellow = yellow + 1

    #move pivot into place
    (A[l], A[yellow - 1]) = (A[yellow - 1], A[l])  

    #then sort the 2 unsorted lists: one before pivot and after pivot
    Quicksort(A, l, yellow-1)
    Quicksort(A, yellow, r)
```

```python
l = list(range(7500, 0, -1))
```
